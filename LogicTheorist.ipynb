```python
import random

# 公理は以下の5つ
  #ソクラテス=>人間
  #人間=>考える葦
  #ソクラテス=>サッカー選手
  #人間=>動物
  #動物=>弱肉強食
  #動物=>死ぬ
axiom_pre=["ソクラテス","人間","ソクラテス","人間","動物","動物"]
axiom_cons=["人間","考える葦","サッカー選手","動物","弱肉強食","死ぬ"]

# 上記の公理をつかって、以下の命題の真偽を確かめます。
# ソクラテス => 死ぬ 
proposition_pre="ソクラテス"
proposition_cons="死ぬ"

X=proposition_pre #Xを公理から探して、見つかったら入れ替える
print(X)

patiance=20 #あきらめるまでの回数です
while True:
  patiance=patiance-1
  flag=0
  for i in range(30):#いろいろ考えてみる回数
    p=random.randint(0,len(axiom_pre)-1) #ランダムに一つ公理を選ぶ
    if axiom_pre[p]==X: #選んだ公理の前提がXに等しければ
      print("=>",axiom_cons[p]) #　プリントして
      X=axiom_cons[p] #Xを選んだ公理の帰結にする
      if X==proposition_cons:# Xが命題の帰結と一致したら
        flag=1
        break
  if flag==0: # いろいろ考えても見つからなかった場合
    if patiance<1: #もし忍耐が0なら
      print("////////////")
      print("あきらめた")
      print("////////////")
      break
    else:
      X=proposition_pre #最初に戻る
      print("---------やりなおし---------")
      print(X)
  else: #見つかってたら終わり
    break
  
```
